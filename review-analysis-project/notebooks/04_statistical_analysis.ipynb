```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/processed/cleaned_reviews.csv")
print("Shape:", df.shape)
df.head()
```

```python
df.describe()
```

```python
corr = df.corr(numeric_only=True)

plt.figure()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
verified = df[df["verified"] == True]["rating"]
non_verified = df[df["verified"] == False]["rating"]

print("Verified Mean:", verified.mean())
print("Non-Verified Mean:", non_verified.mean())
```

```python
t_stat, p_value = ttest_ind(verified, non_verified, equal_var=False)

print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
plt.figure()
sns.boxplot(x="verified", y="rating", data=df)
plt.title("Verified vs Non-Verified Ratings")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="sentiment", y="review_length", data=df)
plt.title("Review Length vs Sentiment")
plt.show()
```

```python
plt.figure()
sns.boxplot(x="sentiment", y="helpful", data=df)
plt.title("Helpful Votes vs Sentiment")
plt.show()
```

```python
plt.figure()
sns.histplot(df["rating"], bins=5, kde=True)
plt.title("Rating Distribution")
plt.show()
```
